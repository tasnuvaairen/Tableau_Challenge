## Dependencies 

```python
import pandas as pd
from sqlalchemy import create_engine
import datetime as dt
import json
import requests
import time
import urllib
#from api_keys import weather_api_key
#from citipy import citipy
import matplotlib.pyplot as plt
```

### Store CSV into DataFrame

```python
# Import citibike CSV file and Read it (Jan-Jun, 2019)


citibike1901_csv = pd.read_csv('201901-citibike-tripdata.csv')
citibike1902_csv = pd.read_csv('201902-citibike-tripdata.csv')
citibike1903_csv = pd.read_csv('201903-citibike-tripdata.csv')
citibike1904_csv = pd.read_csv('201904-citibike-tripdata.csv')
citibike1905_csv = pd.read_csv('201905-citibike-tripdata.csv')
citibike1906_csv = pd.read_csv('201906-citibike-tripdata.csv')
citibike1907_csv = pd.read_csv('201907-citibike-tripdata.csv')
citibike1908_csv = pd.read_csv('201908-citibike-tripdata.csv')
citibike1909_csv = pd.read_csv('201909-citibike-tripdata.csv')
citibike1910_csv = pd.read_csv('201910-citibike-tripdata.csv')
citibike1911_csv = pd.read_csv('201911-citibike-tripdata.csv')
citibike1912_csv = pd.read_csv('201912-citibike-tripdata.csv')

combined_df = pd.concat([citibike1901_csv, citibike1902_csv, citibike1903_csv,citibike1904_csv, citibike1905_csv, 
                         citibike1906_csv, citibike1907_csv, citibike1908_csv, citibike1909_csv, citibike1910_csv, 
                         citibike1911_csv, citibike1912_csv])

combined_df 
```

## Clean CSV data

```python
# Clean the CSV DF Column headers
citibike_2019 = combined_df.rename(columns = {'tripduration':'Trip Duration', 'starttime':'Start Time', 
                                                'stoptime':'Stop Time','start station id':'Start Station ID',
                                                'start station name':'Start Station Name', 'start station latitude':'Start Station Latitude', 
                                                'start station longitude':'Start Station Longitude','end station id':'End Station ID',
                                                'end station name':'End Station Name','end station latitude':'End Station Latitude', 
                                                'end station longitude':'End Station Longitude','bikeid':'Bike ID','usertype':'User Type',
                                                'birth year':'Birth Year', 'gender': 'Gender'})
citibike_2019
```

```python
# Fix the date and time columns
citibike_2019['Start Date']=pd.to_datetime(citibike_2019['Start Time']).dt.strftime('%Y-%m-%d') #.date
citibike_2019['Start Time']=pd.to_datetime(citibike_2019['Start Time']).dt.strftime('%H:%M') #.dt.time
citibike_2019['Stop Date']=pd.to_datetime(citibike_2019['Stop Time']).dt.strftime('%Y-%m-%d') #.date
citibike_2019['Stop Time']=pd.to_datetime(citibike_2019['Stop Time']).dt.strftime('%H:%M') #.dt.time
citibike_2019
```

```python
citibike_2019.to_csv('citibike_2019.csv')
print (citibike_2019)
```
